```python
df['id']=df['id'].fillna(df['id'].mean())
```

```python
df['id'].isnull().sum()
```

```python
sns.boxplot(x='id',data=df)
```

```python

Q1 = df['id'].quantile(0.25)
Q3 = df['id'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['id'] = np.where(df['id'] < lower_bound, lower_bound,
                       np.where(df['id'] > upper_bound, upper_bound, df['id']))
```

```python
sns.boxplot(x='id',data=df)
```
